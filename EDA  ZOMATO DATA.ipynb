```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_data=pd.read_csv('zomato.csv')
```

```python
zomato_data.head()
```

```python
zomato_data.info()
```

```python
# deleting unnecessary column
zomato=zomato_data.drop(['url','dish_liked','phone'],axis=1)
```

```python
zomato.head(10)
```

```python
# removing the duplicates
zomato.duplicated().sum()
zomato.drop_duplicates(inplace=True)
```

```python
#removing null datas
zomato.isnull().sum()
zomato.dropna(how='any',inplace=True)
```

```python
zomato.info()
```

```python
zomato.columns.values
```

```python
# rename column names
zomato=zomato.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
zomato.columns
```

```python
# some transformations
# changing type of cost
zomato['cost']=zomato['cost'].astype(str)
# using lambda to remove ',' from cost
zomato['cost']=zomato['cost'].apply(lambda x: x.replace(',','.'))
# again changing type to float
zomato['cost']=zomato['cost'].astype(float)
```

```python
zomato.info()
```

```python
#reading unique 'rate' dataset
zomato['rate'].unique()
```

```python
# removing '/5' from rate
zomato = zomato.loc[zomato.rate !='NEW']
zomato = zomato.loc[zomato.rate !='-'].reset_index(drop=True)
remove_slash=lambda x:x.replace('/5','') if type(x) == np.str else x
zomato.rate = zomato.rate.apply(remove_slash).str.strip().astype('float')
zomato['rate'].head()
```

```python
zomato.head()
```

```python
zomato.info()
```

```python
zomato.name=zomato.name.apply(lambda x:x.title())
zomato.online_order.replace(('Yes','No'),('True','False'),inplace=True)
zomato.book_table.replace(('Yes','No'),('True','False'),inplace=True)
zomato.cost.unique()
```

```python
def Encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','cost','votes'])]:
        zomato[column]=zomato[column].factorize()[0]
    return zomato
zomato_en=Encode(zomato.copy())
```

```python
zomato_en
```

```python
# get correlation between variables
corr=zomato_en.corr(method='kendall')
plt.figure(figsize=(15,8),dpi=100)
sns.heatmap(corr,annot=True)
zomato_en.columns
```

```python
#the highest relationship is between name and address which is 0.62
```

```python
#DATA VISUALISATION
# restaurants delivering online or not
sns.countplot(zomato['online_order'])
plt.title('restaurants delivering online or not')
```

```python
#restaurant allowing table booking or not
plt.title('restaurant allowing table booking or not')
sns.countplot(zomato['book_table'])
```

```python
#table booking rate vs rate
y=pd.crosstab(zomato['rate'],zomato['book_table'])
y.div(y.sum(1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['red','yellow'])
```

```python
plt.figure(figsize=(15,6),dpi=100)
sns.countplot(zomato['city'])
sns.countplot(zomato['city']).set_xticklabels(sns.countplot(zomato['city']).get_xticklabels(),rotation=90)
plt.title('location')
```

```python
#location vs rating
plt.figure(figsize=(15,5))
loc_plot=pd.crosstab(zomato['rate'],zomato['city'])
loc_plot.plot(kind='bar',stacked=True)
plt.title('location vs rating')
plt.legend(loc=(1.05,1))
```
